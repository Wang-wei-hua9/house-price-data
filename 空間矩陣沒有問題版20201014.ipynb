```python
#創建關係矩陣
# %%
import pandas as pd
import numpy as np
from tqdm import tqdm
import sys
import re

#讀檔
inpath = 'C:\\Users\\lol\\Desktop\\新北市七區研究\\新北市資料七區空間遞延\\板橋空間遞延\\50M\\關係矩陣\\a3 w50M.csv'
df = pd.read_csv(inpath)

#用[]list 命名['i']、['j']、['wij']
df = df[['i','j','wij']]

#將panda的Series形式轉成 list的形式 使用.to_list()
i = df['i'].to_list()
j = df['j'].to_list()
wij = df['wij'].to_list()

#建立函式 get_TableValue(x) 選取的 i,j 並用 mask列出我們需要的 i和 j
#用loc選取整個欄位 並找到我們要的 (i,j) return 那個值
def get_TableValue(x):
    i = x[0]
    j = x[1]
    mask = (df['i']==i) & (df['j']==j)
    target = df.loc[mask]['wij'].to_list()

    return target[0]

#建立一個zeros的矩陣，而且是一個(4639,4639)的矩陣
z = np.zeros([4693, 4693])

#建立迴圈 對x做 zip的元素打包 變成 tuple(類組) 並以(i,j)的形式呈現 因為前面以建成list的模式並不是tuple
#將對應得值扣除第一欄第一列(以變建立成需要的行列) 使這樣的迴圈 呼叫get_TableValue(x) 並執行這項
for x in tqdm(zip(i, j)):
    z[x[0]-1][x[1]-1] = get_TableValue(x)

# %%
```

```python
#檢查矩陣結果
print(z)
```
